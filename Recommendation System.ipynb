# Text Preprocessing

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
df['tags']
```

# pip install nltk

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stemming(word):
    y = []
    for i in word.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
df['tags'].iloc[0]
```

```python
df['tags'] = df['tags'].apply(stemming)
```

```python
df['tags'].iloc[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
vector = cv.fit_transform(df['tags'])
```

```python
vector.shape
```

```python
cv.get_feature_names_out().tolist()
```

# Calculate Similarity Score

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[0]
```

```python
similaritydf = pd.DataFrame(similarity,index=df['title'],columns=df['title'])
```

```python
similaritydf
```

```python
similaritydf.loc['Avatar'].sort_values(ascending=False)[1:6]
```

```python
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:5]
```

```python
df[df["title"]=='Avatar'].index[0]
```

```python
def recommend(movie):
    movie_index=df[df['title']==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:5]
    for i in movie_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
